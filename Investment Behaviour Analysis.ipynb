##### TASK-8 : INVESTMENT DURATION

```python
# Purpose of the Task : To Calculate the Average Investment Duration
```

```python
# To analyze investment duration column
# Specify the column containing investment durations
investment_duration = 'Duration' 
```

```python
# Mapping categorical durations to numerical midpoints
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6  # Adjust this value based on domain knowledge if necessary
}
```

```python
# Replace categorical values with numerical midpoints
Dataset['Duration_Midpoints'] = Dataset[investment_duration].map(duration_mapping)

# Drop rows with missing or NaN values in the mapped column
cleaned_data = Dataset['Duration_Midpoints'].dropna()
cleaned_data
```

```python
# To Display counts for each category
duration_counts = Dataset[investment_duration].value_counts()
print("\n Duration Category Counts")
print(duration_counts)
```

```python
# To Calculate the average investment duration
average_duration = cleaned_data.mean()

# To Display the average investment duration
print("\n Average Investment Duration")
print(f"The estimated average investment duration is {average_duration:.2f} years.")
```

##### TASK-10 : CORRELATION ANALYSIS

```python
# Purpose of the Task : Explore Potential Correlation Between Factors
```

```python
# To select revelent columns
# Select relevant columns (ensure columns are numerical or encoded properly)
selected_columns = ['age', 'Duration', 'Expect']  # Update with actual column names

# Encode categorical columns if necessary (e.g., Duration, Expect)
# Example: Convert 'Duration' to numerical values
duration_mapping = {
    'Less than 1 year': 0.5,
    '1-3 years': 2,
    '3-5 years': 4,
    'More than 5 years': 6
}
Dataset['Duration'] = Dataset['Duration'].map(duration_mapping)

expect_mapping = {
    '10%-20%': 15,
    '20%-30%': 25,
    '30%-40%': 35
}
Dataset['Expect'] = Dataset['Expect'].map(expect_mapping)

# Check for missing values and handle them
print("\nChecking for Missing Values")
print(Dataset[selected_columns].isnull().sum())
```

```python
import seaborn as sns
```

```python
# Calculate correlation coefficients
correlation_matrix = Dataset[selected_columns].corr()
print("\n Correlation Matrix")
print(correlation_matrix)
```

```python
# Visualize the correlation matrix with a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Matrix", fontsize=16)
plt.show()
```

```python
# Scatter plots for pairwise relationships
sns.pairplot(Dataset[selected_columns], diag_kind='kde', kind='scatter', plot_kws={'alpha': 0.6})
plt.suptitle("Pairwise Relationships Between Selected Factors", y=1.02, fontsize=16)
plt.show()
```

```python
# Encoding Gender (0: Female, 1: Male)
Dataset['gender'] = Dataset['gender'].map({'Female': 0, 'Male': 1})
```
